Looking at the visualizations, we see that sodium level is a pretty good indicator of whether a cereal is healthy or not. Thus, we decided to use a logistic regression model on the dataset except for non-numeric data and the feature itself to produce a guess-timate of the sodium level in each cereal. As above, we realized that predicting a value wasn't going to be very accurate, so we decided to provide a range that demonstrates whether a cereal is healthy or not. A cereal, in our case, is healthy if sodium levels are below 110mg and unhealthy if it is above or equal to 110mg. We trained and tested our dataset with 0.25 of the initially randomly arranged cereals data (not sorted like the visualization part). Comparing to the initial data, our logistic regression model predicts, with the generated sodium values, a cereal's healthiness correctly 84.2% of the time. Below, we will try to train all the data from our initial breakfast_cereals dataset and test its accuracy on a different dataset of more cereals. We will first do some preprocessing of the cereal.csv dataset to take the form of the original dataset so we can run the logistic regression on it.
